```
%cd /
from glob import glob

img_list = glob('/content/dataset/guide_dog/export/images/*.jpg')

print(len(img_list))
```

```
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```
with open('content/dataset/guide_dog/train.txt', 'w') as f:
  f.write('\n'.join(train_img_list)+'\n')

with open('content/dataset/guide_dog/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list)+'\n')
```

```
import yaml
with open('/content/dataset/guide_dog/data.yaml', 'r') as f:
  data = yaml.load(f)

print(data)

data['train'] = '/content/dataset/guide_dog/train.txt'
data['val'] = '/content/dataset/guide_dog/val.txt'

with open('/content/dataset/guide_dog/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```
